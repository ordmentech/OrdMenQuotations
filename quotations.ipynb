```python

with open("q.txt") as f:
    quotes = f.read()
    print(repr(quotes))
```

```python
quotes_without_new_line = quotes.replace("\n\n", "\n")
```

```python
quotes_list = quotes_without_new_line.split("\n")
```

```python
quotes_list
```

```python
import pandas as pd

# Define the column names
columns = ['English', 'Author', 'Persian']

# Create an empty DataFrame with these columns
df = pd.DataFrame(columns=columns)
```

```python

for i in range(len(quotes_list)):
    
        
    data = {
        'English': "",
        'Author': "",
        'Persian': ""
    }

    if i % 2 == 0: # even - زوج
        splitted = quotes_list[i].split(": ")
        author = splitted[0]
        english_quote = splitted[1].replace("\"", "")
        if len(english_quote) < 95:
            data["English"] = english_quote
            data['Author'] = author
            splitted = quotes_list[i+1].split(": ")
            persian = splitted[1].replace("\"", "")
            data['Persian'] = persian
            new_row_df = pd.DataFrame([data])
            df = pd.concat([df, new_row_df], ignore_index=True)
```
